```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
raw_data = pd.read_csv("train_FD003.txt", sep=" ", header=None)
```

```python
raw_data
```

```python
raw_data = raw_data.drop([26,27], axis=1)

raw_data.columns = ['Engine_ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 'SensorMeasure13', 'SensorMeasure14', 'SensorMeasure15', 'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 'SensorMeasure19', 'SensorMeasure20', 'SensorMeasure21']

raw_data[raw_data['Engine_ID'] == 1]
```

```python
EOL=[]
for sl_no in raw_data["Engine_ID"]:
    EOL.append(((raw_data[raw_data['Engine_ID']==sl_no]["Cycle"]).values)[-1])
```

```python
EOL
```

```python
raw_data["EOL"]=EOL
```

```python
raw_data["LR"] = raw_data["Cycle"].div(raw_data["EOL"])
```

```python
raw_data
```

```python
raw_data = raw_data.drop(["Engine_ID","EOL"], axis =1)
```

```python
labels = []

for i in range (0,len(raw_data)):
    if np.array(raw_data["LR"])[i] <= 0.6:
        labels.append(0)
    elif np.array(raw_data["LR"])[i] <=0.8:
        labels.append(1)
    else:
        labels.append(2)
        
raw_data["labels"]=labels

raw_data = raw_data.drop(["LR"], axis=1)
```
