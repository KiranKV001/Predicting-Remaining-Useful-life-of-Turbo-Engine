### Classification Problem Formulation 

The challange of this competetion was to predict the Remaining Useful Life of the engine by using the given sensor's data and operational conditions.

But in this project, we will try to simplify that by converting into classification probelm. where the classes will be three types.
i.e. <b>Good Condition</b>, <b>Moderate Condition</b> and <b>Warning Condition</b>.

Labels corresponding to each conditions.

  * Good Condition - 0
  * Moderate Condition - 1
  * Warning Condition - 3


### Defining the labels

Here will define the engine's condition with <b>Life Ratio(LR)</b>, which is the ratio between <b>Current Cycle</b>. and the <b>End Cycle/ End Of Cycle(EOL)<b/>.
    
if <b>LR=0</b>, that component just started its degradation and <b>LR=1</b> means it is completely degraded.

* if LR<= 0.6 - Good condition
* if 0.6<LR<= 0.8 - Moderate condition
* if 0.8<LR<= 0.6 - Good condition


```python
# calculate Life Ration(LR)

raw_data["LR"] = raw_data["Cycle"].div(raw_data["EOL"])
```

```python
raw_data
```

```python
# we create LifeRatio(LR) using End of Cycle based Engine_ID due that drop these columns.

raw_data = raw_data.drop(["Engine_ID","EOL"], axis =1)
```

```python
raw_data
```

```python
labels = []

for i in range (0,len(raw_data)):
    if np.array(raw_data["LR"])[i] <= 0.6:
        labels.append(0)
    elif np.array(raw_data["LR"])[i] <=0.8:
        labels.append(1)
    else:
        labels.append(2)
        
raw_data["labels"]=labels

raw_data = raw_data.drop(["LR"], axis=1)
```

### Spliting the data into training and testing

```python
X=dtrain.drop(['labels'],axis=1)
y=dtrain['labels']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.30, random_state = 1)
```

```python
print("Training Dataset X_train: ",X_train.shape)
print("Training Dataset y_train: ",y_train.shape)
print("Training Dataset X_test: ",X_test.shape)
print("Training Dataset y_test: ",y_test.shape)
```

### Training Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier

rf_clas = RandomForestClassifier()

rf_clas.fit(X_train,y_train)
```

### Performance on Validation set

```python
y_pred_rf = rf_clas.predict(X_test)
```

```python
y_pred_rf
```

```python
from sklearn import metrics
```

```python
cm = pd.DataFrame(confusion_matrix(y_test, y_pred_rf, normalize ='true')*100)
colormap = sns.color_palette("Reds")
sns.heatmap(cm, annot=True,cmap=colormap)

accuracy_score(y_test, y_pred_rf)
```

### Randonmized search with cross validation for Hyperparameter optimization

```python
from sklearn.model_selection import RandomizedSearchCV

# Number of tress in random forest
n_estimators = [200,300,400]

# Maximum number of levels in tree
max_depth = [10,20,30]

# Max number of samples given to each tree
max_samples = [0.2,0.3,0.4]

# Create the random grid
random_grid = {
    'n_estimators': n_estimators,
    'max_depth': max_depth,
    'max_samples': max_samples
}

print(random_grid)
```

```python
rf_random = RandomizedSearchCV(estimator = rf_clas, param_distributions = random_grid, cv = 3, n_jobs = -1)
```

```python
rf_random.fit(X_train,y_train)
```
