### Classification Problem Formulation 

The challange of this competetion was to predict the Remaining Useful Life of the engine by using the given sensor's data and operational conditions.

But in this project, we will try to simplify that by converting into classification probelm. where the classes will be three types.
i.e. <b>Good Condition</b>, <b>Moderate Condition</b> and <b>Warning Condition</b>.

Labels corresponding to each conditions.

  * Good Condition - 0
  * Moderate Condition - 1
  * Warning Condition - 2


### Defining the labels

Here will define the engine's condition with <b>Life Ratio(LR)</b>, which is the ratio between <b>Current Cycle</b>. and the <b>End Cycle/ End Of Cycle(EOL)<b/>.
    
if <b>LR=0</b>, that component just started its degradation and <b>LR=1</b> means it is completely degraded.

* if LR<= 0.6 - Good condition
* if 0.6<LR<= 0.8 - Moderate condition
* if 0.8<LR<= 1 - Warning condition


```python
# calculate Life Ration(LR)

raw_data["LR"] = raw_data["Cycle"].div(raw_data["EOL"])
```

```python
raw_data
```

```python
# we create LifeRatio(LR) using End of Cycle based Engine_ID due that drop these columns.

raw_data = raw_data.drop(["Engine_ID","EOL"], axis =1)
```

```python
raw_data
```

```python
labels = []

for i in range (0,len(raw_data)):
    if np.array(raw_data["LR"])[i] <= 0.6:
        labels.append(0)
    elif np.array(raw_data["LR"])[i] <=0.8:
        labels.append(1)
    else:
        labels.append(2)
        
raw_data["labels"]=labels

raw_data = raw_data.drop(["LR"], axis=1)
```
